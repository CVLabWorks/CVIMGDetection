# All Models Performance Visualization

This notebook visualizes performance across all four models: ResNet50, BNN, Softmax, and BNN+Attention.

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

# Set style for elegant plots with white background
plt.style.use('default')
sns.set_palette("husl")
```

```python
# Dataset names
datasets = ['ADM', 'BigGAN', 'GLIDE', 'Midjourney', 'SD v1.5', 'VQDM', 'Wukong']

# Define elegant colors
color_acc = '#2E86AB'        # Deep blue for overall accuracy
color_real = '#A23B72'       # Purple for real images
color_fake = '#F18F01'       # Orange for fake images

# BNN Model Data
bnn_acc = [89.00, 90.80, 88.30, 71.20, 79.30, 67.90, 81.70]
bnn_real_acc = [98.80, 100.00, 98.80, 61.00, 76.60, 56.00, 79.40]
bnn_fake_acc = [79.20, 81.60, 77.80, 81.40, 82.00, 79.80, 84.00]

# Softmax Model Data
softmax_acc = [89.20, 89.70, 88.10, 80.80, 85.40, 81.00, 83.80]
softmax_real_acc = [78.80, 79.40, 76.60, 77.20, 79.20, 81.80, 80.00]
softmax_fake_acc = [99.60, 100.00, 99.60, 84.40, 91.60, 80.20, 87.60]

# BNN+Attention Model Data
bnn_att_acc = [91.30, 92.60, 90.30, 77.90, 87.90, 82.30, 86.30]
bnn_att_real_acc = [97.00, 99.60, 98.60, 72.20, 91.40, 79.80, 86.00]
bnn_att_fake_acc = [85.60, 85.60, 82.00, 83.60, 84.40, 84.80, 86.60]

# ResNet Model Data (for comparison)
resnet_acc = [90.10, 90.00, 89.60, 80.80, 85.20, 78.80, 87.20]
resnet_real_acc = [81.00, 80.80, 79.60, 78.00, 77.60, 79.00, 81.20]
resnet_fake_acc = [99.20, 99.20, 99.60, 83.60, 92.80, 78.60, 93.20]
```

## 1. BNN Model Performance

```python
# Create BNN plot
fig, ax = plt.subplots(figsize=(12, 7))

x_pos = np.arange(len(datasets))

# Plot the three lines with markers
line1 = ax.plot(x_pos, bnn_acc, marker='o', linewidth=2.5, markersize=8, 
                color=color_acc, label='Overall Accuracy', alpha=0.9)
line2 = ax.plot(x_pos, bnn_real_acc, marker='s', linewidth=2.5, markersize=8, 
                color=color_real, label='Real Image Accuracy', alpha=0.9)
line3 = ax.plot(x_pos, bnn_fake_acc, marker='^', linewidth=2.5, markersize=8, 
                color=color_fake, label='Fake Image Accuracy', alpha=0.9)

# Customize the plot
ax.set_xlabel('Dataset Type', fontsize=13, fontweight='bold')
ax.set_ylabel('Accuracy (%)', fontsize=13, fontweight='bold')
ax.set_title('BNN Model Performance Across Different AI-Generated Image Datasets', 
             fontsize=15, fontweight='bold', pad=20)

# Set x-axis
ax.set_xticks(x_pos)
ax.set_xticklabels(datasets, fontsize=11)

# Set y-axis range
ax.set_ylim(50, 105)
ax.set_yticks(range(50, 105, 5))
ax.tick_params(axis='y', labelsize=11)

# Add grid with white background
ax.set_facecolor('white')
fig.patch.set_facecolor('white')
ax.grid(True, alpha=0.5, linestyle='-', linewidth=0.6, color='#CCCCCC')
ax.set_axisbelow(True)

# Add legend
ax.legend(loc='lower left', fontsize=11, framealpha=0.9, edgecolor='gray')

# Add horizontal reference line at 80%
ax.axhline(y=80, color='gray', linestyle=':', linewidth=1, alpha=0.5)

plt.tight_layout()
plt.savefig('bnn_performance.png', dpi=300, bbox_inches='tight')
plt.show()

print(f"BNN Model - Average Overall Accuracy: {np.mean(bnn_acc):.2f}%")
print(f"BNN Model - Average Real Accuracy: {np.mean(bnn_real_acc):.2f}%")
print(f"BNN Model - Average Fake Accuracy: {np.mean(bnn_fake_acc):.2f}%")
```

```python
# Heatmap comparison
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

models_data = [
    ('ResNet50', [resnet_acc, resnet_real_acc, resnet_fake_acc]),
    ('BNN', [bnn_acc, bnn_real_acc, bnn_fake_acc]),
    ('Softmax', [softmax_acc, softmax_real_acc, softmax_fake_acc]),
    ('BNN+Attention', [bnn_att_acc, bnn_att_real_acc, bnn_att_fake_acc])
]

for idx, (model_name, data) in enumerate(models_data):
    ax = axes[idx // 2, idx % 2]
    
    # Create heatmap data
    heatmap_data = np.array(data)
    
    # Create heatmap
    im = ax.imshow(heatmap_data, cmap='RdYlGn', aspect='auto', vmin=50, vmax=100)
    
    # Set ticks
    ax.set_xticks(np.arange(len(datasets)))
    ax.set_yticks(np.arange(3))
    ax.set_xticklabels(datasets, fontsize=10)
    ax.set_yticklabels(['Overall Acc', 'Real Acc', 'Fake Acc'], fontsize=10)
    
    # Rotate x labels
    plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")
    
    # Add values
    for i in range(3):
        for j in range(len(datasets)):
            text = ax.text(j, i, f'{heatmap_data[i, j]:.1f}',
                          ha="center", va="center", color="black", fontsize=9, fontweight='bold')
    
    ax.set_title(f'{model_name} Performance Heatmap', fontsize=12, fontweight='bold', pad=10)
    
    # Add colorbar
    cbar = plt.colorbar(im, ax=ax)
    cbar.set_label('Accuracy (%)', fontsize=10)

fig.patch.set_facecolor('white')
plt.tight_layout()
plt.savefig('all_models_heatmap.png', dpi=300, bbox_inches='tight')
plt.show()
```

## Key Findings

1. **Best Overall Model**: BNN+Attention achieves the highest average accuracy (87.6%)
2. **Most Consistent**: Softmax shows balanced real/fake detection
3. **Challenge Dataset**: Midjourney and VQDM remain difficult for all models
4. **BNN Characteristics**: High real accuracy but lower fake accuracy, indicating conservative predictions
5. **ResNet/Softmax**: Excellent fake detection (>90%) but moderate real detection (~79%)
